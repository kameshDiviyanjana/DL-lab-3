```python
# First install all required packages

# Import libraries
import os
from dotenv import load_dotenv
from google.generativeai import configure  # Critical for Colab
from langchain_google_genai import ChatGoogleGenerativeAI, GoogleGenerativeAIEmbeddings
from langchain_community.document_loaders import PyPDFLoader, UnstructuredPowerPointLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.vectorstores import FAISS
from langchain.chains import RetrievalQA

# Load environment variables (create this cell first!)
from google.colab import userdata
gemini_api_key = userdata.get('GEMINI_API_KEY')  # Recommended Colab method

# Alternative: Use .env file (upload it first)
# from google.colab import files
# files.upload()  # Upload your .env file
# load_dotenv()
# gemini_api_key = os.getenv("GEMINI_API_KEY")

# Configure authentication (solves the metadata error)
configure(api_key=gemini_api_key)

# 1. Load documents (modify path to your Colab folder)
documents = []
folder_path = "/content/sample_data/lectet"  # Create this folder and upload files

# # Create the folder if it doesn't exist
# !mkdir -p "{folder_path}"

# Upload your files (run this cell separately if needed)
# from google.colab import files
# uploaded = files.upload()
# for filename in uploaded.keys():
#   !mv "{filename}" "{folder_path}/"

# Process files
for file in os.listdir(folder_path):
  file_path = os.path.join(folder_path, file)
  if file.lower().endswith(".pdf"):
      loader = PyPDFLoader(file_path)
      documents.extend(loader.load())
  elif file.lower().endswith((".ppt", ".pptx")):
      loader = UnstructuredPowerPointLoader(file_path)
      documents.extend(loader.load())

# 2. Split documents
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200
)
texts = text_splitter.split_documents(documents)

# 3. Create vector store
embeddings = GoogleGenerativeAIEmbeddings(
    model="models/embedding-001",
    model_name="models/embedding-001",
    google_api_key=gemini_api_key
)
db = FAISS.from_documents(texts, embeddings)
# embeddings = GoogleGenerativeAIEmbeddings(
#     model="models/embedding-001",
#     google_api_key=gemini_api_key  # Still required for LangChain
# )
# db = FAISS.from_documents(texts, embeddings)

# 4. Initialize LLM
# llm = ChatGoogleGenerativeAI(
#     model="gemini-pro",
#     temperature=0,
#     google_api_key=gemini_api_key,
#     convert_system_message_to_human=True
# )
llm = ChatGoogleGenerativeAI(
    model="gemini-1.5-pro-latest",  # Updated model name
    temperature=0,
    google_api_key=gemini_api_key,
    convert_system_message_to_human=True
)

# 5. Create QA chain
qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=db.as_retriever(search_kwargs={"k": 3}),
    return_source_documents=True,
    verbose=True
)

# 6. Query function
def ask_question(question):
    result = qa_chain({"query": question})
    answer = result["result"]
    sources = list(set(
        os.path.basename(doc.metadata["source"])
        for doc in result["source_documents"]
    ))
    return {"answer": answer, "sources": sources or ["General knowledge"]}

# Test it
response = ask_question("Explain transformer architecture")
print("Answer:", response["answer"])
print("Sources:", ", ".join(response["sources"]))
```

```python
import ipywidgets as widgets
from IPython.display import display

question_input = widgets.Text(
    placeholder='Type your question here...',
    description='Question:',
    layout=widgets.Layout(width='600px')
)

# Create a button
submit_button = widgets.Button(
    description='Ask',
    button_style='success'  # 'success', 'info', 'warning', 'danger' or ''
)

# Create an output area
output = widgets.Output()

# Define what happens when button is clicked
def on_button_clicked(b):
    with output:
        output.clear_output()
        print("Thinking... ü§î")
        import time
        time.sleep(0.8)
        response = ask_question(question_input.value)
        output.clear_output()
        print("‚úÖ Here's what I found:")
        print("\nüìù Answer:", response["answer"])
        print("\nüîó Sources:", ", ".join(response["sources"]))

# Connect button to function
submit_button.on_click(on_button_clicked)

# Show the widgets
display(question_input, submit_button, output)
```
